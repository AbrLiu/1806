## Problem 4

**About Stoichiometric Matrices**


[ Quick Summary for those
who do not want to read everything: chemical reactions turn into matrices by turning 
coefficients into matrix elements. Every reaction is a column in the matrix. The coefficients on the left side of the
chemical reaction appear with negative signs.
The nullspace corresponds to the reactions' steady state with negative signs
corresponds to backwards reactions.]


Suppose that you have a system of chemical reactions. You know the velocity, or rate of each individual reaction, and you know which species are involved in each reaction. You can think of the reactants and reaction rates as the vertices and edges in a network, or graph. We create something very similar to the incidence matrix from a graph, and use it to discover the steady state solutions to a system of reactions.

<b> Basics of chemical reactions </b>

Recall the following basics about chemical reactions.

If A is a molecule, then we use [A] to denote the concentration of that molecular species. The concentration is a function of time, but we typically are interested in the concentration at steady state (no time dependence).

If A→B is a unimolecular reaction, then the velocity or reaction rate, v, is written as

$$ v=\frac{d[B]}{dt}=−\frac{d[A]}{dt}=k[A].$$
If A+B→P+Q is a bimolecular chemical reaction, then the reaction rate is

$$v=\frac{d[P]}{dt}=\frac{d[Q]}{dt}=−\frac{d[A]}{dt}=−\frac{d[B]}{dt}=k[A][B].$$

<b> Example 1: Simple Stoichiometric Matrix </b>

Consider the single reaction with three different chemical species.

2NO+O2⇌2NO2.
<br>

The stoichiometric matrix is a column vector:

$$\begin{pmatrix} 2 \\ -1 \\ 2 \end{pmatrix}$$
such that

 [[NO] [O$_2$] [NO$_2$]] $\begin{pmatrix} 2 \\ -1 \\ 2 \end{pmatrix} =0. $


Note the convention that reactants have negative sign, and products have positive sign.

<b> Example 2: Input and output </b>

Now consider the same reaction that is no longer closed. We will allow that oxygen is being continuously added into the system, and we are taking out nitrogen dioxide. The reaction is now the following system:

r1: 2NO+O$_2$⇌2NO$_2$ <br>
e1: &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; →O$_2$ <br>
e2: &nbsp;&nbsp;&nbsp;NO$_2$→



                

The stoichiometric matrix is


$\begin{matrix} & r_1 & e_1 &  e_2 \\
NO & -2 & 0 & 0 \\
O_2 & -1 & 1 & 0 \\
NO_2 & 2 & 0 & -1
\end{matrix}$




 	 	 	 
<b> Example 3: Closed System of Reactions </b>

Suppose we have the following closed system of chemical reactions. It is closed because no new reactants are entering the system, and no products are leaving the system.

r1: H$_2$O + CO ⇌CO$_2$  + H$_2$<br>
r2: H$_2$O + H  ⇌ H$_2$ + OH <br>
r3: OH + CO   ⇌ CO$_2$  + H




The  following molecular species can be designated:

 X$_1$	= H$_2$O	 	
 X$_2$	= CO	 	 
 X$_3$	= CO$_2$	  
 X$_4$	= H$_2$	 	 
 X$_5$	= H	 	 
 X$_6$	= OH	 	
    
Then the coefficient on every term in this reaction is 1, the entries will all be +1, −1, or 0. The convention is that the reactants have negative, and the products have positive sign.

The stoichiometric matrix is given by the following, where the columns each designate a particular reaction, and the coefficients determine the amount of each species present.

$\begin{matrix} & r_1 & e_1 &  e_2 \\
X_1=H_20 & -1 & -1 & 0 \\
X_2=CO & -1 & 0 & -1 \\
X_3=CO_2  & 1 & 0 &  1 \\
X_4 = H_2 & 1 & 1 & 0 \\
X_5 =H & 0 & -1 & 1 \\
X_6 = OH & 0 & 1 & -1
\end{matrix}$



Let $v_1$ is the velocity of reaction 1, $v_2$ is the velocity of reaction 2, and $v_3$ is the velocity of reaction 3, then

$$\begin{pmatrix}
\frac{d[X_1]}{dt} \\
\frac{d[X_2]}{dt} \\
\frac{d[X_3]}{dt} \\
\frac{d[X_4]}{dt} \\
\frac{d[X_5]}{dt} \\
\frac{d[X_6]}{dt} 
\end{pmatrix}=
\begin{pmatrix}
-1 & -1 & 0 \\
-1 & 0 & -1 \\
1 & 0 & 1\\
1 & 1 & 0 \\
0 & -1 & 1 \\
0 & 1 & -1
\end{pmatrix}
\begin{pmatrix}
v_1 \\
v_3 \\
v_3
\end{pmatrix}.
$$

To find the steady state solution, we are interested in finding the velocities such that each concentration is steady, that is the derivative is zero:

$$\begin{pmatrix}
0 \\ 0 \\ 0 \\
0 \\ 0 \\ 0 
\end{pmatrix}=
\begin{pmatrix}
-1 & -1 & 0 \\
-1 & 0 & -1 \\
1 & 0 & 1\\
1 & 1 & 0 \\
0 & -1 & 1 \\
0 & 1 & -1
\end{pmatrix}
\begin{pmatrix}
v_1 \\
v_3 \\
v_3
\end{pmatrix}.
$$

This is exactly the nullspace of the matrix. In this case the nullspace, which corresponds to the steady state is:

$$\begin{pmatrix}
v_1 \\
v_3 \\
v_3
\end{pmatrix}=
\begin{pmatrix}
-1 \\
1 \\
1
\end{pmatrix} .
$$
This is telling us that at steady state, the first reaction will be proceeding in the backwards direction, and the other two in the forwards direction, all with the same velocity, or reaction rate k.
